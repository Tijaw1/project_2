# Data Cleaning

# Data Encoding

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
# Label each column
le.fit(app_df["CODE_GENDER"])
app_df["CODE_GENDER"] = le.transform(app_df["CODE_GENDER"])

le.fit(app_df["FLAG_OWN_CAR"])
app_df["FLAG_OWN_CAR"] = le.transform(app_df["FLAG_OWN_CAR"])

le.fit(app_df["FLAG_OWN_REALTY"])
app_df["FLAG_OWN_REALTY"] = le.transform(app_df["FLAG_OWN_REALTY"])

le.fit(app_df["NAME_INCOME_TYPE"])
app_df["NAME_INCOME_TYPE"] = le.transform(app_df["NAME_INCOME_TYPE"])

le.fit(app_df["NAME_EDUCATION_TYPE"])
app_df["NAME_EDUCATION_TYPE"] = le.transform(app_df["NAME_EDUCATION_TYPE"])

le.fit(app_df["NAME_FAMILY_STATUS"])
app_df["NAME_FAMILY_STATUS"] = le.transform(app_df["NAME_FAMILY_STATUS"])

le.fit(app_df["NAME_HOUSING_TYPE"])
app_df["NAME_HOUSING_TYPE"] = le.transform(app_df["NAME_HOUSING_TYPE"])

le.fit(app_df["OCCUPATION_TYPE"])
app_df["OCCUPATION_TYPE"] = le.transform(app_df["OCCUPATION_TYPE"])

app_df.head()
```

```python
# clean credit_record data
credit_df.set_index(credit_df['ID'], inplace=True)
```

```python
credit_df.drop(columns=['ID'], inplace=True)
credit_df.head()
```

```python
credit_df.info()
```

```python
# group application and credit records dataframes
joined = pd.merge(app_df, credit_df, left_on='ID', right_on='ID')
joined.shape
```

```python
joined.head()
```

```python
# encode status column
le.fit(joined["STATUS"])
joined["STATUS"] = le.transform(joined["STATUS"])
```

```python
joined.head()
```

```python
# Save data for future use
joined_encoded_path = Path("/Users/aalfiler/Sites/project_2/Data/joined_encoded.csv")
joined_df.to_csv(joined_encoded_path, index=False)
```

# Determine Feature Importance

```python
# Define features set
X = joined.drop(columns="STATUS")
y = joined["STATUS"]
```

```python
# Check the balance of our target values
y.value_counts()
```

```python
# Split the X and y into X_train, X_test, y_train, y_test
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# Create the StandardScaler instance
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
# Fit the Standard Scaler with the training data
X_scaler = scaler.fit(X_train)
```

```python
# Scale the training and testing data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Resample the training data with the BalancedRandomForestClassifier
from imblearn.ensemble import BalancedRandomForestClassifier

joined_rf = BalancedRandomForestClassifier(n_estimators=100, random_state=1)
joined_rf.fit(X_train, y_train)
```
